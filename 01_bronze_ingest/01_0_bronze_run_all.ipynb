```python
from notebookutils import mssparkutils
from datetime import datetime

RUN_ID = datetime.utcnow().strftime("%Y%m%d_%H%M%S")

bronze_notebooks = [
    "01_1_bronze_business",
    "01_2_bronze_review",
    "01_3_bronze_checkin",
    "01_4_bronze_user",
    "01_5_bronze_tip"    
]

results = []

print(f"[START] Bronze orchestration started. RUN_ID = {RUN_ID}")

for nb in bronze_notebooks:
    print(f"[RUN] Starting notebook: {nb}")

    try:
        result = mssparkutils.notebook.run(
            nb,
            600,
            {"batch_id": RUN_ID}
        )

        print(f"[DONE] {nb} finished. Return = {result}")
        results.append((nb, "SUCCESS", result))

    except Exception as e:
        print(f"[FAIL] {nb} failed. Error = {str(e)}")
        results.append((nb, "FAILED", str(e)))
        raise

print("\n[SUMMARY]")
for nb, status, msg in results:
    print(f"{nb}: {status} | {msg}")

print(f"\n[END] Bronze orchestration finished. RUN_ID = {RUN_ID}")
```
