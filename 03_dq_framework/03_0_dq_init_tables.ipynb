```python
from pyspark.sql.types import (
    StructType,StructField,
    StringType,IntegerType,LongType,DoubleType,TimestampType
)

DB = "yelp_lakehouse.dbo"
```

```python
# dq_run_log
dq_run_log_schema = StructType([
    StructField("dq_run_id", StringType(),False),
    StructField("pipeline_run_id", StringType(),True),
    StructField("layer", StringType(),False),
    StructField("table_name", StringType(), False),
    StructField("run_ts", TimestampType(), False),
    StructField("end_ts", TimestampType(), True),
    StructField("total_rules", IntegerType(), True),
    StructField("passed_rules", IntegerType(), True),
    StructField("failed_rules", IntegerType(), True),
    StructField("critical_failed_rules", IntegerType(), True),
    StructField("major_failed_rules", IntegerType(), True),
    StructField("final_decision", StringType(), True), # PASS / DEGRADED / BLOCKED
    StructField("status", StringType(), True),   # SUCCESS / FAILED
    StructField("error_message", StringType(), True)
])

(
    spark.createDataFrame([], dq_run_log_schema).write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(f"{DB}.dq_run_log")
)
```

```python
# 2) dq_rule_result
dq_rule_result_schema = StructType([
    StructField("dq_run_id", StringType(), False),
    StructField("pipeline_run_id", StringType(), True),
    StructField("run_ts",TimestampType(), False),
    StructField("layer", StringType(), False),
    StructField("table_name", StringType(), False),
    StructField("rule_id", StringType(), False),
    StructField("rule_name", StringType(), False),
    StructField("rule_type", StringType(),True),
    StructField("column_name", StringType(),True),
    StructField("severity", StringType(), False),
    StructField("dimension", StringType(), False),
    StructField("total_count", LongType(), True),
    StructField("failed_count",LongType(), True),
    StructField("failed_rate", DoubleType(), True),
    StructField("threshold_rate", DoubleType(), True),
    StructField("status", StringType(), False),   # PASS / FAIL
    StructField("rule_message", StringType(), True)
])

(
    spark.createDataFrame([],dq_rule_result_schema).write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(f"{DB}.dq_rule_result")
)
```

```python
# 3) dq_table_gate
dq_table_gate_schema = StructType([
    StructField("dq_run_id", StringType(), False),
    StructField("pipeline_run_id", StringType(), True),
    StructField("run_ts", TimestampType(), False),
    StructField("layer", StringType(), False),
    StructField("table_name", StringType(),False),
    StructField("total_rules", LongType(), True),
    StructField("passed_rules", LongType(),True),
    StructField("failed_rules", LongType(), True),
    StructField("critical_failed_rules", LongType(), True),
    StructField("major_failed_rules", LongType(), True),
    StructField("decision", StringType(), False),   # PASS / DEGRADED / BLOCKED
    StructField("decision_reason", StringType(), True)
])

(
    spark.createDataFrame([], dq_table_gate_schema).write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(f"{DB}.dq_table_gate")
)
```
