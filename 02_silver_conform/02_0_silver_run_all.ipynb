```python
from notebookutils import mssparkutils
from datetime import datetime

try:
    batch_id = mssparkutils.widgets.get("batch_id")
except Exception:
    batch_id = datetime.utcnow().strftime("%Y%m%d_%H%M%S")

silver_notebooks = [
    "02_1_silver_business",
    "02_2_silver_review",
    "02_3_silver_checkin",
    "02_4_silver_user",
    "02_5_silver_tip"
]

results = []

print(f"[START] Silver orchestration started. batch_id = {batch_id}")

for nb in silver_notebooks:
    print(f"[RUN] Starting notebook: {nb}")

    try:
        result = mssparkutils.notebook.run(
            nb,
            600,
            {"batch_id": batch_id}
        )

        print(f"[DONE] {nb} finished. Return = {result}")
        results.append((nb, "SUCCESS", result))
    except Exception as e:
        print(f"[FAIL] {nb} failed. Error = {str(e)}")
        results.append((nb, "FAILED", str(e)))

        raise
print("\n[SUMMARY]")

for nb, status, msg in results:
    print(f"{nb}: {status} | {msg}")

print(f"\n [END] Silver orchestration finished. batch_id = {batch_id}")
```
