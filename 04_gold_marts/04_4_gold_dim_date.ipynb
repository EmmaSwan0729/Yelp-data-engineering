```python
from pyspark.sql import functions as F

GOLD = "dim_date"

df_date = spark.table("review_silver")

df_dim = (
    df_date.select("date")
    .dropna()
    .dropDuplicates()
    .withColumn("date_id", F.date_format("date","yyyyMMdd").cast("int"))
    .withColumn("year", F.year("date"))
    .withColumn("quarter", F.quarter("date"))
    .withColumn("month", F.month("date"))
    .withColumn("day", F.dayofyear("date"))
    .withColumn("weekofyear", F.weekofyear("date"))
    .withColumn("dayofweek", F.dayofweek("date"))
    .withColumn("day_name", F.date_format("date","E"))
    .select(
        "date_id",
        "date",
        "year",
        "quarter",
        "month",
        "day",
        "weekofyear",
        "dayofweek",
        "day_name"
    )
)

(
    df_dim.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(GOLD)

)
```
